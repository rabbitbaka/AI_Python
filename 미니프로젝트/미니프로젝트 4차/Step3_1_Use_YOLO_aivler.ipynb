#### 1) Label 파일의 클래스 값 변경

```python
import os
import glob

# 반복할 하위 폴더 목록
subfolders = ['test', 'train', 'valid']

# 폴더 목록

# 각 폴더와 하위 폴더의 labels 파일을 처리
for subfolder in subfolders:
    labels_folder = os.path.join(other_extract_folder, subfolder, 'labels')

    # labels 폴더의 모든 txt 파일 가져오기
    label_files = glob.glob(os.path.join(labels_folder, '*.txt'))

    # 각 txt 파일에서 띄어쓰기 전의 숫자를 1로 변경
    for label_file in label_files:
        with open(label_file, 'r') as f:
            lines = f.readlines()

        # 새로운 내용 생성
        new_lines = []
        for line in lines:
            # 줄에서 첫 번째 숫자만 1로 변경
            parts = line.split(' ')
            if parts:  # 비어 있지 않은 줄인 경우
                parts[0] = '1'  # 첫 번째 숫자 변경
                new_line = ' '.join(parts)  # 변경된 내용을 다시 조합
                new_lines.append(new_line)
            else:
                new_lines.append(line)  # 비어 있는 줄은 그대로 유지

        # 변경된 내용을 파일에 다시 저장
        with open(label_file, 'w') as f:
            f.writelines(new_lines)

        print(f'{label_file} 파일의 첫 번째 숫자가 1로 변경되었습니다.')

print('모든 txt 파일의 첫 번째 숫자 변경 작업이 완료되었습니다.')
```
