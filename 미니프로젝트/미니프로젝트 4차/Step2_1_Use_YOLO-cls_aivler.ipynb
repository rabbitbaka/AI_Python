### (2) 데이터셋 전처리

* **세부 요구사항**
    - 데이터셋을 전처리 합니다.
        - YOLO-cls 모델이 요구하는 폴더 구조를 만듭니다.
            1. Datasets라는 폴더를 생성합니다.
            2. Training set, Validation set, Test set(선택 사항) 각 데이터셋이 들어갈 폴더를 생성합니다.
            3. 각 데이터셋 폴더에 분류할 클래스의 이름을 가진 폴더를 생성합니다.
        - 폴더 구조에 맞게 데이터를 분배합니다.
    - 예시 코드에서 사용한 라이브러리
        - os, glob, random, shutil, numpy

```python
import glob
import random
import shutil
import numpy as np
```

#### 1) 모델이 요하는 구조의 폴더 생성

```python

# Datasets파일 안에 train, val, test 파일생성
data_folder = '/content/Datasets'
train_data = '/content/Datasets/train'
val_data = '/content/Datasets/val'
test_data = '/content/Datasets/test'

if not os.path.exists(data_folder):
    os.makedirs(data_folder)

if not os.path.exists(train_data):
    os.makedirs(train_data)

if not os.path.exists(val_data):
    os.makedirs(val_data)

if not os.path.exists(test_data):
    os.makedirs(test_data)

print(os.listdir(data_folder))
```

```python
# my, class로 나눌 파일 생성
class_names = ['my', 'other']
for cn in class_names :
  temp = os.path.join(train_data, cn)
  if not os.path.exists(temp):
    os.makedirs(temp)
    print(os.path.exists(temp))
for cn in class_names :
  temp = os.path.join(val_data, cn)
  if not os.path.exists(temp):
    os.makedirs(temp)
    print(os.path.exists(temp))
for cn in class_names :
  temp = os.path.join(test_data, cn)
  if not os.path.exists(temp):
    os.makedirs(temp)
    print(os.path.exists(temp))
```

#### 2) 각 폴더에 이미지 데이터 옮기기

```python
img_list_my = sorted(glob.glob(my_extract_folder + '/*'))
img_list_other = sorted(glob.glob(other_extract_folder + '/*'))
len(img_list_my), len(img_list_other)
```

```python
# 섞기
random.seed(2024)
random.shuffle(img_list_my)
random.shuffle(img_list_other)
```

```python
## Test set의 비율 설정
test_size = 0.2
val_size = 0.2

## 나의 얼굴 파일 리스트와 다른 얼굴 파일 리스트에 대한 반복문
for i_l in [img_list_my, img_list_other] :
    ## 리스트의 길이 체크
    list_len = len(i_l)
    #list_len = 11000  ## 예시 파일의 이미지 갯수가 맞지 않아서 11000개까지만 사용
    ## 데이터 분할을 위한 인덱스 설정
    split_idx = int(list_len * (1 - test_size) )
    split_idx_2 = int(split_idx * (1 - val_size) )

    ## 인덱스를 이용해 상위 리스트를 Training set, Test set, Val set 3가지로 세분화
    list_tr = i_l[ : split_idx_2]
    list_val = i_l[split_idx_2 : split_idx]
    list_te = i_l[split_idx : list_len]

    ## 현재 리스트가 나의 얼굴 파일 리스트와 같다면
    if i_l == img_list_my :
        ## "나의 얼굴 파일 리스트"의 파일을 Training set 폴더 안의 "나의 얼굴 폴더"로 복사
        ## 이동이 잘못 되었을 경우를 생각하여 복사로 진행
        for file_path in list_tr :
            f_name = file_path.split('/')
            shutil.copy(src=file_path,
                        dst=train_data+'/my/'+f_name[-1]
                        )
            print(f'파일 이동 완료 : {f_name[-1]}')

        ## "나의 얼굴 파일 리스트"의 파일을 Validation set 폴더 안의 "나의 얼굴 폴더"로 복사
        ## 이동이 잘못 되었을 경우를 생각하여 복사로 진행
        for file_path in list_val:
            f_name = file_path.split('/')
            shutil.copy(src=file_path, dst=val_data+'/my/'+f_name[-1])
            print(f'파일 이동 완료 : {f_name[-1]}')

        ## "나의 얼굴 파일 리스트"의 파일을 Test set 폴더 안의 "나의 얼굴 폴더"로 복사
        ## 이동이 잘못 되었을 경우를 생각하여 복사로 진행
        for file_path in list_te :
            f_name = file_path.split('/')
            shutil.copy(src=file_path,
                        dst=test_data+'/my/'+f_name[-1]
                        )
            print(f'파일 이동 완료 : {f_name[-1]}')

    ## 현재 리스트가 "나의 얼굴 파일 리스트"가 아니라면 (즉, "다른 사람 얼굴 파일 리스트"라면)
    else :
        ## "다른 사람 얼굴 파일 리스트"의 파일을 Training set 폴더 안의 "다른 사람 얼굴 폴더"로 복사
        ## 이동이 잘못 되었을 경우를 생각하여 복사로 진행
        for file_path in list_tr :
            f_name = file_path.split('/')
            shutil.copy(src=file_path,
                        dst=train_data+'/other/'+f_name[-1],
                        )
            print(f'파일 이동 완료 : {f_name[-1]}')

        ## "다른 사람 얼굴 파일 리스트"의 파일을 Validation set 폴더 안의 "다른 사람 얼굴 폴더"로 복사
        ## 이동이 잘못 되었을 경우를 생각하여 복사로 진행
        for file_path in list_val:
            f_name = file_path.split('/')
            shutil.copy(src=file_path, dst=val_data+'/other/'+f_name[-1])
            print(f'파일 이동 완료 : {f_name[-1]}')

        ## "다른 사람 얼굴 파일 리스트"의 파일을 Test set 폴더 안의 "다른 사람 얼굴 폴더"로 복사
        ## 이동이 잘못 되었을 경우를 생각하여 복사로 진행
        for file_path in list_te :
            f_name = file_path.split('/')
            shutil.copy(src=file_path,
                        dst=test_data+'/other/'+f_name[-1]
                        )
            print(f'파일 이동 완료 : {f_name[-1]}')
```

```python
print('Training data의 my_face 이미지 수 : ', len(os.listdir(train_data+'/my')))
print('Training data의 other_face 이미지 수 : ', len(os.listdir(train_data+'/other')))

print('Validation data의 my_face 이미지 수 : ', len(os.listdir(val_data+'/my')))
print('Validation data의 other_face 이미지 수 : ', len(os.listdir(val_data+'/other')))

print('Test data의 my_face 이미지 수 : ', len(os.listdir(test_data+'/my')))
print('Test data의 other_face 이미지 수 : ', len(os.listdir(test_data+'/other')))
```
