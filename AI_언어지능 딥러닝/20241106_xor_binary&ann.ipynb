```python
import torch
x_train = torch.FloatTensor([[0,0],[0,1],[1,0],[1,1]])
y_train = torch.FloatTensor([[0],[1],[1],[0]])
W = torch.randn([2,1], requires_grad=True)
b = torch.randn(1, requires_grad=True)
optimizer = torch.optim.SGD([W,b], lr=0.01)

# [딥러닝 1단계] 모델을 만든다. Model Setup
def H(x):
  model = torch.sigmoid(torch.matmul(x,W) +b)
  return model

# [딥러닝 2단계] 학습을 시킨다. Training - W,b찾기
for i in range(200000):
  cost = torch.mean((-1)*y_train*torch.log(H(x_train))+(-1)*(1-y_train)*torch.log(1-H(x_train)))
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()
  if i%10000 == 0: print('[Iteration]',i,'[Cost]',cost.detach().item())
# [딥러닝 3단계] 추론을 수행한다. Inference
model_result = H(x_train)
print(model_result.detach())
```

```python
import torch
x_train = torch.FloatTensor([[0,0],[0,1],[1,0],[1,1]])
y_train = torch.FloatTensor([[0],[1],[1],[0]])
W_h = torch.randn([2,3], requires_grad=True)
b_h = torch.randn([3], requires_grad=True)
W_o = torch.randn([3,1], requires_grad=True)
b_o = torch.randn([1], requires_grad=True)
optimizer = torch.optim.SGD([W_h,b_h,W_o,b_o], lr=0.01)

# [딥러닝 1단계] 모델을 만든다. Model Setup
def H(x):
  hl1 = torch.sigmoid(torch.matmul(x,W_h)+b_h) # 첫번째 히든레이어
  model = torch.sigmoid(torch.matmul(hl1,W_o)+b_o) # 두번째 히든레이어
  return model

# [딥러닝 2단계] 학습을 시킨다. Training - W,b찾기
for i in range(200000):
  cost = torch.mean((-1)*y_train*torch.log(H(x_train))+(-1)*(1-y_train)*torch.log(1-H(x_train)))
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()
  if i%10000 == 0: print('[Iteration]',i,'[Cost]',cost.detach().item())

# [딥러닝 3단계] 추론을 수행한다. Inference
model_result = H(x_train)
print(model_result.detach())
```
