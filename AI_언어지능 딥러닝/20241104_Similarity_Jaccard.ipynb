```python
import numpy as np
from sklearn.metrics import accuracy_score

print(accuracy_score(np.array([1,3,2]),np.array([1,4,5])))
print(accuracy_score(np.array([1,3,2]),np.array([4,1,5])))
print(accuracy_score(np.array([1,1,0,0]),np.array([1,1,0,2])))
print(accuracy_score(np.array([1,0,1,0]),np.array([1,1,0,2])))
```

```python
import numpy as np
from sklearn.metrics import jaccard_score

print(jaccard_score(np.array([1,3,2]),np.array([1,4,5]), average='weighted'))
print(jaccard_score(np.array([1,3,2]),np.array([4,1,5]), average='weighted'))
print(jaccard_score(np.array([1,3,2]),np.array([2,1,3]), average='weighted'))
print(jaccard_score(np.array([1,1,0,0]),np.array([1,1,0,2]), average='weighted'))
print(jaccard_score(np.array([1,0,1,0]),np.array([1,1,0,2]), average='weighted'))
```
