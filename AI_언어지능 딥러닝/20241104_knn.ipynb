```python
import numpy as np
import random

group_A = np.random.uniform(100, 200, (15, 4)) #앞의 값은 샘플 수, 샘플의 디멘션(특징)
group_B = np.random.uniform(0, 50, (15, 4))
group_C = np.column_stack((np.random.uniform(100, 200, 15), np.random.uniform(0, 50, 15), np.zeros((15, 2))))

# 각 그룹의 점들과 우리가 가지고 있는 현재 Point와의 거리를 구하는 함수 필요
def distance(p, g):
    return np.sqrt(((g - p) ** 2).sum(axis=1))

point = np.array([0, 0, 0, 0])
distances_A = distance(point, group_A)
distances_B = distance(point, group_B)
distances_C = distance(point, group_C)

k = 5
distances_A = np.sort(distances_A)[:k]
distances_B = np.sort(distances_B)[:k]
distances_C = np.sort(distances_C)[:k]

# 평균 구하기
distances_A = distances_A.mean()
distances_B = distances_B.mean()
distances_C = distances_C.mean()

# 각 그룹의 통계치를 나타내는 리스트
stat = [distances_A, distances_B, distances_C]  # 제일 앞은 k개 중에 몇개가 A에 가장 가까운지에 대한 수치
# 이 stat에서 최대로 큰 값의 위치를 구하면 그게 그룹이 됨.
index = np.argmin(stat)
if index == 0:
    print("Group A")
elif index == 1:
    print("Group B")
else:
    print("Group C")
```
