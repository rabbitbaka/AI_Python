## 1. 히든레이어가 1개 있는 구조

```python
# 히든레이어가 있는 구조
import torch
x_train = torch.FloatTensor([[1,2],[2,3],[3,4],[4,4],[5,3],[6,2],[8,1]])
y_train = torch.FloatTensor([[0],[0],[0],[1],[1],[1],[0]])
nH1 = 100
W_h = torch.randn([2,nH1], requires_grad=True)
b_h = torch.randn([nH1], requires_grad=True)
W_o = torch.randn([nH1,1], requires_grad=True)
b_o = torch.randn([1], requires_grad=True)
optimizer = torch.optim.SGD([W_h,b_h,W_o,b_o], lr=0.01)

# [딥러닝 1단계] 모델을 만든다. Model Setup
def H(x):
  # 히든레이어 만들기
  H1 = torch.sigmoid(torch.matmul(x,W_h)+b_h)
  model = torch.sigmoid(torch.matmul(H1,W_o)+b_o)
  return model

# [딥러닝 2단계] 학습을 시킨다. Training - w,b찾기
for i in range(100000):
  cost = torch.mean((-1)*y_train*torch.log(H(x_train))+(-1)*(1-y_train)*torch.log(1-H(x_train)))
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()
  if i%10000 == 0: print('[Iteration]',i,'[Cost]',cost.detach().item())
# [딥러닝 3단계] 추론을 수행한다. Inference
x_test = torch.FloatTensor([7,1])
model_result = H(x_train)
print('Model Inference with [7,1]:')
print(model_result.detach())
```

## 2. 히든레이어가 3개 있는 구조

```python
# 히든레이어가 3개가 있는 구조
import torch
x_train = torch.FloatTensor([[1,2],[2,3],[3,4],[4,4],[5,3],[6,2],[8,1]])
y_train = torch.FloatTensor([[0],[0],[0],[1],[1],[1],[0]])
nH1 = 10
nH2 = 5
nH3 = 8
# 히든레이어 1 : 10개
W_h1 = torch.randn([2,nH1], requires_grad=True)
b_h1 = torch.randn([nH1], requires_grad=True)
# 히든레이어 2 : 5개
W_h2 = torch.randn([nH1,nH2], requires_grad=True)
b_h2 = torch.randn([nH2], requires_grad=True)
# 히든레이어 3 : 8개
W_h3 = torch.randn([nH2,nH3], requires_grad=True)
b_h3 = torch.randn([nH3], requires_grad=True)
# 아웃풋레이어
W_o = torch.randn([nH3,1], requires_grad=True)
b_o = torch.randn([1], requires_grad=True)
optimizer = torch.optim.SGD([W_h1,b_h1,W_h2,b_h2,W_h3,b_h3,W_o,b_o], lr=0.01)

# [딥러닝 1단계] 모델을 만든다. Model Setup
def H(x):
  # 히든레이어 3개 연결하기
  H1 = torch.sigmoid(torch.matmul(x,W_h1)+b_h1)
  H2 = torch.sigmoid(torch.matmul(H1,W_h2)+b_h2)
  H3 = torch.sigmoid(torch.matmul(H2,W_h3)+b_h3)
  model = torch.sigmoid(torch.matmul(H3,W_o)+b_o)
  return model

# [딥러닝 2단계] 학습을 시킨다. Training - w,b찾기
for i in range(100000):
  cost = torch.mean((-1)*y_train*torch.log(H(x_train))+(-1)*(1-y_train)*torch.log(1-H(x_train)))
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()
  if i%10000 == 0: print('[Iteration]',i,'[Cost]',cost.detach().item())
# [딥러닝 3단계] 추론을 수행한다. Inference
x_test = torch.FloatTensor([7,1])
model_result = H(x_train)
print('Model Inference with [7,1]:')
print(model_result.detach())
```
