### 2) 모델 구조만 빌려와서 사용해보기 (권장하지 않음)

* 모델 구조 선택하기

```python
model_scratch = YOLO(model='yolo11n.pt', task='detect')
```

* 모델 학습하기

```python
results_train = model_scratch.train(model='/content/yolo11n.pt',
                                    data='/content/rock-paper-scissors-14/data.yaml',
                                    epochs=1,
                                    patience=5,
                                    seed=2024,
                                    pretrained=False,
                                    )
```
