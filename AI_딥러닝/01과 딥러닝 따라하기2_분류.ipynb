# 1.환경준비

* 라이브러리 Import

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import *
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

# 2.Classification : 대학원 지원

## (1) 데이터 전처리

### 1) 데이터 준비

```
path = "https://raw.githubusercontent.com/DA4BAM/dataset/master/Graduate_apply.csv"
data = pd.read_csv(path)
data.head()
```

```
target = 'admit'
x = data.drop(target, axis=1)
y = data.loc[:, target]
```

### 2) 가변수화

### 3) 데이터분할

```
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=.2, random_state = 20)
```

## (2) 머신러닝 모델링 : 로지스틱회귀

```
# 필요한 알고리즘을 불러 옵시다.
from sklearn.linear_model import LogisticRegression
```

### 1) 모델 선언

```
model = LogisticRegression()
```

### 2) 학습

```
model.fit(x_train, y_train)
```

### 3) 예측

```
pred = model.predict(x_val)
```

### 4) 검증
만든 모델은 얼마나 정확한지 검증해 봅시다.



```
print(confusion_matrix(y_val, pred))
print('-'*50)
print(classification_report(y_val, pred))
```

## (3) 딥러닝 모델링
* 필요한 함수들 불러오기
* 모델 선언
* 학습
* 예측
* 성능 검증

### 1) 전처리 : Scaling

```
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_val = scaler.transform(x_val)
```

### 2) 필요한 함수 불러오기

```
from keras.models import Sequential
from keras.layers import Dense, Input
from keras.backend import clear_session
```

### 3) 모델 선언

```
nfeatures = x_train.shape[1] #num of columns
nfeatures
```

```
# 메모리 정리
clear_session()

# Sequential 타입 모델 선언
model = Sequential([Input(shape = (nfeatures,)),
                    Dense(1, activation = 'sigmoid' ) ])

# 모델요약
model.summary()
```

```
model.compile(optimizer='adam', loss= 'binary_crossentropy')
```

### 4) 학습

```
model.fit(x_train, y_train)
```

### 5) 예측

```
pred = model.predict(x_val)
pred = np.where(pred>= 0.5, 1, 0)
```

### 6) 검증
만든 모델은 얼마나 정확한지 검증해 봅시다.



```
print(confusion_matrix(y_val, pred))
print('-'*50)
print(classification_report(y_val, pred))
```

# 3.Classification : mobile

## (1) 데이터 전처리

### 1) 데이터 준비

```
path = "https://raw.githubusercontent.com/DA4BAM/dataset/master/mobile_churn_simple.csv"
data = pd.read_csv(path)
data['CHURN'] = data['CHURN'].map({'STAY':0, 'LEAVE':1})
data.head()
```

|	구분	|	변수 명	|	내용	|	type	|	비고	|
|	----	|	----	|	----	|	----	|	----	|
|	**Target**	|	**CHURN**	|	이탈여부	|	범주	| 0,1	|
|	feature	|	INCOME	|	소득수준(달러)	|	숫자	|		|
|	feature	|	OVERAGE	|	월평균 초과사용시간(분)	|	숫자	| |
|	feature	|	LEFTOVER	|	월평균 잔여시간(%)	|	숫자	| 	|
|	feature	|	HOUSE	|	집가격(달러)	|	숫자	|	|
|	feature	|	HANDSET_PRICE	|	휴대폰가격(달러)	|	숫자	|		|
|	feature	|	OVER_15MINS_CALLS_PER_MONTH	|	월평균 장기통화 횟수	|	숫자	| 		|
|	feature	|	AVERAGE_CALL_DURATION	|	평균통화시간(분)	|	숫자	|		|

```
target = 'CHURN'
x = data.drop(target, axis=1)
y = data.loc[:, target]
```

### 2) 가변수화

### 3) 데이터분할

```
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=.2, random_state = 20)
```

### 4) Scaling

```
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_val = scaler.transform(x_val)
```

## (2) 모델링
* 필요한 함수들 불러오기
* 모델 선언
* 학습
* 예측
* 성능 검증

```
from keras.models import Sequential
from keras.layers import Dense, Input
from keras.backend import clear_session
```

### 1) 모델 선언

```
nfeatures = x_train.shape[1] #num of columns
nfeatures
```

```
# 메모리 정리
clear_session()

# Sequential 타입 모델 선언
model = Sequential([Input(shape = (nfeatures,)),
                    Dense(1, activation = 'sigmoid' ) ])

# 모델요약
model.summary()
```

```
model.compile(optimizer='adam', loss='binary_crossentropy')
```

### 2) 학습

```
model.fit(x_train, y_train)
```

### 3) 예측

```
pred = model.predict(x_val)
pred
```

```
pred = np.where(pred>= 0.5, 1, 0)
```

### 4) 검증
만든 모델은 얼마나 정확한지 검증해 봅시다.



```
print(confusion_matrix(y_val, pred))
print('-'*50)
print(classification_report(y_val, pred))
```
